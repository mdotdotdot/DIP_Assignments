```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('AssignmentPhoto.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print("Current picture resolution:", image.shape)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
image_resized = cv2.resize(image,(256,256))
print("Resized image resolution:",image_resized.shape)
plt.imshow(image_resized)
plt.axis('off')
plt.show()
```

```python
grey_image = cv2.cvtColor(image_resized,cv2.COLOR_BGR2GRAY)
print("Resultion image:", grey_image.shape)
plt.imshow(grey_image, cmap = "gray")
plt.axis('off')
plt.show()
```

```python
_,binary_image = cv2.threshold(grey_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
# cv2.threshold() returns a tuple containing two values: the thresholded image and a return code
# unpacking the second value (which is the binary image) from the tuple
plt.imshow(binary_image, cmap = "gray")
plt.axis('off')
plt.show()
```

```python
_,binary_image = cv2.threshold(grey_image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
contours, _ = cv2.findContours(binary_image,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
min_area = 20
coins = []
for contour in contours:
  area = cv2.contourArea(contour)
  if area > min_area:
    coins.append(contour)
print("Number of Coins Identified: ",len(coins))
cv2.drawContours(binary_image, coins, -1, (255, 255, 255), -1)
plt.imshow(binary_image)
plt.axis('off')
plt.show()
```
